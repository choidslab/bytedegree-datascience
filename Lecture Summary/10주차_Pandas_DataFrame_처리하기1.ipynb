### DataFrame Boolean Selection으로 데이터 선택하기
  - 특정 조건에 맞는 row만 선택

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('./train.csv')
```

```python
train_data.head()
```

```python
pclass = train_data['Pclass'] == 1 # 조건식으로 조건에 맞는 데이터만 select 가능
age = (train_data['Age'] >= 30) & (train_data['Age'] < 40) 

# 30대이면서 1등석에 탄 사람
pclass & age # boolean 값으로 결과 출력
train_data[pclass & age] # train_data에 boolean 결과를 넣으면 해당 조건에 맞는 데이터 결과를 출력
```

### DataFrame에 새 column(컬럼) 추가/삭제하기
  - 기존 컬럼으로부터 새로운 컬럼 생성 또는 삭제
  - `[]`를 이용하여 추가
  - `insert 함수`를 이용하여 원하는 위치에 추가

#### 새로운 Column 추가하기

```python
train_data['Age_Double'] = train_data['Age'] * 2
train_data.head() # 새롭게 추가되는 column은 기존 DataFrame의 가장 끝에 추가된다.
```

```python
train_data['Age_Tripple'] = train_data['Age_Double'] + train_data['Age']
train_data.head()
```

```python
# insert 함수를 이용하여 원하는 위치에 새로운 column 추가하기
# train_data.insert(3, 'Fare10', train_data['Fare'] / 10) # 기존 Name 컬럼 앞에 Fare10이라는 컬럼을 추가(Fare 컬럼 값을 10으로 나눈 결과)
train_data.head()
```

#### Column 삭제하기
  - `drop 함수`를 이용하여 삭제
  - 리스트를 사용하여 여러개의 컬럼 동시 삭제 가능
  - `axis`를 반드시 지정

```python
train_data.drop(['Age_Tripple'], axis=1)
```

```python
train_data.head() # drop을 이용하여 삭제한 경우 복사본에 대한 컬럼을 삭제한 것이므로 Age_Tripple 컬럼은 그대로 남아 있다.
```

```python
train_data.drop(['Age_Tripple', 'Age_Double'], axis=1) # 2개의 컬럼 동시에 삭제하기
```

```python
train_data.head() # 원본은 변경되지 않음
```

```python
# 원본 데이터를 삭제하기, inplace 파라미터 값을 True로 지정
train_data.drop(['Age_Tripple'], axis=1, inplace=True)
```

```python
train_data.head() # 원본 데이터에서 Age_Tripple 컬럼이 삭제된 것을 확인할 수 있음
```

### DataFrame Column간 상관관계 계산
  - 상관관계: 특정 변수의 값이 변함에 따라 다른 변수에 미치는 영향을 나타낸 것
  - `corr` 함수를 이용하여 상관관계 계산(-1 ~ 1 사이의 결괏값 출력)
    - 1 또는 -1에 가까울 수록 두 변수간의 값이 비슷하게 증감한다. 
    - 0에 가까우면 두 변수간에는 상관관계가 거의 없다는 것을 의미
  - `연속성(숫자형) 있는 데이터`에 대해서만 연산

```python
train_data.corr()
```

```python
plt.matshow(train_data.corr()) # 색이 밝을수록 관계가 깊다는 것을 의미
```

### DataFrame NaN 데이터 처리하기

#### NaN 값 확인
  - `info` 함수를 이용하여 NaN 개수 확인
  - `isna` 함수를 통해 boolean 타입으로 확인

```python
import pandas as pd
```

```python
train_data = pd.read_csv('./train.csv')
train_data.head()
```

```python
train_data.info() # Age, Cabin 컬럼에서 데이터 소실이 있음을 확인할 수 있다.
```

```python
train_data.isna() # True: NaN 값을 의미
```

```python
train_data['Age'].isna() # 특정 컬럼에 대해서만 isna() 함수를 적용할 수도 있음
```

####  NaN 값 처리 방법
  - 데이터에서 삭제하는 방법
    - `dropna` 함수
  - 다른 값으로 치환하여 처리
    - `fillna` 함수

```python
# NaN 데이터 삭제
train_data.dropna() # row에서 NaN 값이 있으면 해당 row 전체 삭제(Column을 삭제하는 것이 아니라 Row를 삭제하는 것을 기억!)
```

```python
train_data.dropna(subset=['Age'])
```

  - `subset`으로 지정된 컬럼 값이 NaN이면 해당 row 삭제 
  - 위의 결과에서 0번째 row를 살펴보면 Cabin 컬럼이 NaN이지만 subset으로 Age를 지정하였고, Age는 22.0이라는 정상 값이 존재하기 때문에 row를 삭제하지 않은 것을 확인할 수 있음
  - subset에 여러개의 컬럼을 지정할 수도 있음

```python
# 여러개의 컬럼을 subset으로 지정
train_data.dropna(subset=['Age', 'Cabin'])
```

```python
# dropna() 함수에서 axis=1로 지정할 경우 NaN이 존재하는 컬럼 전체를 삭제
train_data.dropna(axis=1)
```

#### NaN 값 대체하기
  - 평균으로 대체
  - 생존자/사망자별 평균 대체(타이타닉 데이터 예제)

```python
# 특정 컬럼의 평균 구하기
train_data['Age'].mean()
```

```python
# NaN 값을 컬럼의 평균으로 채워 넣기(대체하기)
train_data['Age'].fillna(train_data['Age'].mean()) # row 888은 원래 NaN -> Age 컬럼의 평균 값 29.699로 변경된 것을 확인
```

```python
# 생존자 나이 평균
survive_mean = train_data[train_data['Survived'] == 1]['Age'].mean()
print(survive_mean)
```

```python
# 사망자 나이 평균
death_mean = train_data[train_data['Survived'] == 0]['Age'].mean()
print(death_mean)
```

```python
train_data[train_data['Survived'] == 1]['Age'].fillna(survive_mean) # 생존자 평균 나이로 NaN을 치환
```

```python
train_data[train_data['Survived'] == 0]['Age'].fillna(death_mean) # 사망자 평균 나이로 NaN을 치환
```

```python
# 실제 데이터 값에 적용하기
train_data.loc[train_data['Survived'] == 1, 'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(survive_mean)
train_data.loc[train_data['Survived'] == 0, 'Age'] = train_data[train_data['Survived'] == 0]['Age'].fillna(death_mean)
```

```python
train_data
```

```python
train_data[train_data['Age'] == survive_mean] # 생존자 평균 나이로 대체된 row 값들 확인
```

```python
train_data[train_data['Age'] == death_mean] # 사망자 평균 나이로 대체된 row 값들 확인
```

```python
# 생존 남성 나이 평균
train_data[(train_data['Survived'] == 1) & (train_data['Sex'] == 'male')]['Age'].mean()
```

```python
# 생존 여성 나이 평균
train_data[(train_data['Survived'] == 1) & (train_data['Sex'] == 'female')]['Age'].mean()
```
