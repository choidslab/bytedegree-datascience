### DataFrame 숫자 데이터/범주형 데이터
  - info 함수를 이용하여 데이터 타입 확인 가능
  - 타입 변경은 `astype` 함수 이용

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('./train.csv')
train_data.info() # info 함수로 데이터 타입 확인
```

#### 숫자형 데이터(Numerical Type)
  - 연속성을 띄는 숫자로 이루어진 데이터
    - ex) Age, Fare 등

```python
train_data['Age']
```

#### 범주형 데이터(Categorical Type)
  - 연속적이지 않은 값(대부분의 경우 숫자형 데이터를 제외한 값)
  - 숫자형 데이터라도 특정 범주를 나타내는 경우 범주형 데이터에 속함
    - ex) Pclass -> 1, 2, 3 숫자형 데이터로 1등급~3등급의 범주를 나타내기 때문에 범주형 데이터에 속함

```python
train_data['Pclass']
```

### 숫자형 데이터를 범주형 데이터로 변환하기

#### Pclass 변환하기
  - astype 함수를 이용하여 타입 변환

```python
train_data
```

```python
train_data.info() # Pclass 데이터 타입: int64
```

```python
train_data['Pclass'] = train_data['Pclass'].astype(str) # Pclass 데이터 타입을 string으로 변환
```

```python
train_data.info() # Pclass의 데이터 타입이 기존 int64 -> string으로 변환된 것을 확인할 수 있음
```

#### Age 숫자형 데이터를 범주형 데이터로 변환하기
  - 변환 logic을 함수로 작성
  - `apply` 함수를 이용하여 적용

```python
def age_int_to_categorical(age):
    if math.isnan(age): # NaN 처리
        return -1
    return math.floor(age / 10) * 10
```

```python
train_data
```

```python
train_data['Age'].apply(age_int_to_categorical)
```

### 범주형 데이터를 숫자형 데이터로 변환하기
  - `One-Hot Encoding`

```python
import pandas as pd
```

```python
train_data = pd.read_csv('./train.csv')
train_data.head()
```

#### One-Hot Encoding
  - 데이터가 특정 범주에 속할 경우 1, 속하지 않을 경우 0으로 표현하는 방법
  - `pandas.get_dummies` 함수를 이용
    - `drop_first`: 첫 번째 카테고리는 사용하지 않음

```python
pd.get_dummies(train_data['Pclass']) # 각 row가 어떤 Pclass에 속하는지 One-Hot Encoding으로 표현, 범주(1, 2, 3)
```

```python
pd.get_dummies(train_data, columns=['Pclass'])
```

```python
pd.get_dummies(train_data, columns=['Pclass'], drop_first=True)
```

### DataFrame `group by`
  - 특정 컬럼을 기준으로 그룹화 
  - SQL group by 절과 동일
  - 데이터 분할/병합, 집계연산

```python
import pandas as pd

data_frame = pd.read_csv('./train.csv')
data_frame.head()
```

```python
class_group = data_frame.groupby('Pclass')
class_group
```

#### groupby의 groups 속성
  - 각 그룹과 그룹에 속한 index를 dict 형태로 반환

```python
class_group.groups # Pclass는 1, 2, 3의 그룹이 존재 -> dict의 key 값
```

```python
gender_group = data_frame.groupby('Sex')
gender_group.groups
```

#### groupping 함수
  - 그룹 데이터에 적용할 수 있는 집계함수(집계함수 적용 시, NaN 값은 집계에서 제외)
    - sum: 데이터의 합
    - count: 데이터 개수
    - mean, std, var: 평균/표준편차/분산
    - min, max: 최소/최대

```python
class_group.count()
```

```python
class_group.sum()
```

```python
class_group.mean()['Age'] # 좌석등급별 평균
```

```python
class_group.mean()['Survived'] # 좌석등급별 생존률
```

```python
data_frame.groupby('Sex').mean()['Survived'] # 성별 생존률
```

#### 여러개의 Column을 선택하여 그룹화하기
  - groupby에 column을 리스트 형태로 전달
  - 집계함수를 적용한 결과는 multi index를 갖는 dataframe

```python
data_frame.groupby(['Pclass', 'Sex']).mean() # 좌석등급과 성별로 그룹화한 결과의 평균 값을 출력
```

```python
data_frame.groupby(['Pclass', 'Sex']).mean().index
```

```python
data_frame.groupby(['Pclass', 'Sex']).mean().loc[(2, 'female')] # 특정 row 검색
```

---

#### index를 이용한 groupby
  - index가 있는 경우, groupby에 level 사용 가능
    - level은 index의 depth를 의미
  - `set_index` 함수
    - Column 데이터를 index 레벨로 변경
  - `reset_index` 함수
    - index 초기화

```python
data_frame.head()
```

```python
data_frame.set_index('Pclass') # Pclass를 index로 이용
```

```python
data_frame.set_index(['Pclass', 'Sex']) # Multi Index
```

```python
data_frame.set_index(['Pclass', 'Sex']).reset_index() # index 초기화 
```

```python
data_frame.set_index('Age').groupby(level=0).mean() # Age가 첫 번째 index가 되고 그룹화를 수행한 뒤, 평균 값을 구함
```

```python
def age_to_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10 # 연령대 구하는 공식
```

```python
# 나이대별 생존률 구하기
data_frame.set_index('Age').groupby(age_to_categorize).mean()
```

#### Multiindex를 이용한 grouping

```python
data_frame.head()
```

```python
data_frame.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).mean()
```

#### 집계(Aggregate) 함수

```python
data_frame.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).aggregate([np.mean, np.sum, np.max]) # numpy를 이용한 집계함수 활용 가능
```
