#### DataFrame 생성하기
  - 대부분의 데이터 분석 업무환경에서 인위적으로 데이터를 생성하는 경우는 많지 않음 
  - 데이터 소스(csv, database, 로그 파일 등)을 활용하여 데이터 분석 업무 수행
  - 본 실습에서는 인위적으로 데이터를 생성(dummy data)

```python
import pandas as pd
```

```python
# DataFrame 생성
data = {'a': 100, 'b': 200, 'c': 300} # key: 속성(column), value: 값
pd.DataFrame(data, index=['x', 'y', 'z']) # index 생성
```

```python
data = {'a': [1, 2, 3], 'b': [2, 3, 4], 'c': [4, 5, 6]} # key: 속성(column), value: 값
pd.DataFrame(data, index=['x', 'y', 'z']) # index 생성
```

#### Series를 이용하여 데이터 생성
  - Series의 인덱스 -> DataFrame의 컬럼

```python
data1 = pd.Series([100, 200, 300, 400], ['a', 'b', 'c', 'd'])
data2 = pd.Series([110, 220, 330, 440], ['a', 'b', 'c', 'd'])
data3 = pd.Series([500, 600, 700, 800], ['a', 'b', 'c', 'd'])
pd.DataFrame([data1, data2, data3])
```

```python
pd.DataFrame([data1, data2, data3], index=[100, 101, 102]) # index를 임의로 지정 가능
```

### DataFrame에서 원하는 column만 선택하는 방법

#### column 선택하기
  - 기본적으로 []는 column 추출
  - 컬럼을 인덱스로 사용할 경우 인덱스 리스트 기능 사용 가능
    - 리스트를 전달하는 경우: DataFrame
    - 하나의 컬럼명을 전달하는 경우: Series

```python
import numpy as np
import pandas as pd
train_data = pd.read_csv('./train.csv')
train_data.head()
```

```python
train_data['Survived'] # 하나의 컬럼명을 넣는 경우 Series 형태로 반환
```

```python
train_data['Name']
```

```python
# 리스트로 여러 인덱스를 지정할 경우, 여러 컬럼 데이터를 가져올 수 있으며 DataFrame 형태로 반환
train_data[['Survived', 'Age', 'Name']] 
```

### DataFrame에서 원하는 row(데이터 값)만 선택하는 방법

#### DataFrame slicing

```python
train_data[:10] # slicing 할 때는 컬럼명으로 지정하지 않아도 됨
```

```python
train_data[1:5]
```

#### row 선택하기
  - Series의 경우 []로 row 선택 가능, **DataFrame의 경우 기본적으로 Column을 선택**하도록 설계되어 있음
  - `.loc`, `.iloc`로 row 선택 가능
    - loc: 인덱스 자체를 사용
    - iloc: 0 based index로 사용
    - 해당 함수들은 `,`를 사용하여 column 선택 가능

```python
train_data.head()
```

```python
train_data.info()
```

```python
train_data.index = np.arange(100, 991)
```

```python
train_data.tail()
```

```python
train_data.loc[989] # Series 형태로 값 출력, Column: index
```

```python
type(train_data.loc[989])
```

```python
# 여러개의 row 값을 가져올 때는 파라미터로 리스트를 전달
train_data.loc[[986, 988]]
```

```python
type(train_data.loc[[986, 988]]) # 여러개의 데이터를 가져올 경우 DataFrame 형태로 출력
```

```python
train_data.iloc[0] # 존재하는 index 값에 관계 없이 base offset 시작을 내부적으로 0으로 처리
```

```python
train_data.iloc[[0, 1, 3]]
```

#### row, column 동시에 선택하기
  - loc, iloc 속성을 이용할 때, 콤마(,)를 이용하면 row, column을 동시에 설정할 수 있음

```python
train_data.loc[[986, 100, 110], ['Survived', 'Name', 'Age', 'Sex']]
```

#### 대괄호 []는 column을 가져올 때 쓰이고, row를 가져올 때는 loc(), iloc()을 사용!
