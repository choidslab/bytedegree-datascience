  - SQL group by 절과 동일
  - 데이터 분할/병합, 집계연산

```python
import pandas as pd

data_frame = pd.read_csv('./train.csv')
data_frame.head()
```

```python
class_group = data_frame.groupby('Pclass')
class_group
```

  - Pclass 3: 439.154786	0.242363	25.140620	0.615071	0.393075	13.675550

#### pivot_table
  - pivot 함수와 기능적으로 동일
  - 중복데이터에 대한 처리 가능
    - 중복되는 값이 있는 경우, aggregation 함수를 사용하여 값을 채워 넣을 수 있음

```python
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '월요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
})
```

```python
# 중복허용 후 호출
df.pivot_table('요일', '지역') # 서울, 월요일 값이 2개로 중복
```

```python
# 서울, 월요일이 2개로 중복이었는데 pivot_table을 이용할 경우 aggregation 함수를 이용하여 평균 값 등을 임의로 넣어줄 수 있다.
pd.pivot_table(df, index='요일', columns='지역', aggfunc=np.mean)
```

### stack, unstack 함수
#### stack & unstack
  - stack: 컬럼 레벨에서 인덱스 레벨로 dataframe 구조 변경
  - unstack: 인덱스 레벨에서 컬럼 레벨로 dataframe 구조 변경

```python
import numpy as np
import pandas as pd

# sample dataframe
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
})
```

```python
df
```

```python
new_df = df.set_index(['지역', '요일']) # multi-index
new_df
```

```python
new_df.unstack(0)
```

```python
new_df.unstack(1)
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```

### Concat 함수를 이용한 DataFrame 병합
  - SQL에서의 Join과 유사한 기능
    - 단순히 2개 이상의 DataFrame을 합쳐주는 기능
    - 기준 없이 그냥 합치는 기능
  - `pandas.concat` 함수
    - axis 지정에 따라 행 또는 열 기준으로 병합 가능(axis=0 : 행단위 병합)

#### column명이 동일한 2개의 DataFrame 생성

```python
df1 = pd.DataFrame({'key1': np.arange(10), 'value1': np.random.randn(10)})
```

```python
df2 = pd.DataFrame({'key1': np.arange(10), 'value1': np.random.randn(10)})
```

```python
df1, df2
```

```python
pd.concat([df1, df2]) # 두 데이터를 행을 기준으로 병합
```

```python
# ignore_index=True로 지정할 경우 index가 새롭게 생성된다.
pd.concat([df1, df2], ignore_index=True)
```

```python
# 열을 기준으로 병합
pd.concat([df1, df2], axis=1)
```

#### column명이 다른 경우

```python
df1 = pd.DataFrame({'key1': np.arange(10), 'value1': np.random.randn(10)})
df3 = pd.DataFrame({'key2': np.arange(10), 'value2': np.random.randn(10)})
```

```python
pd.concat([df1, df3])
```

```python
pd.concat([df1, df3], axis=1)
```

### Merge/Join 함수를 이용한 DataFrame 병합하기
#### merge
  - SQL의 Join과 같이 column을 기준으로 병합
    - join 방식: `how` 파라미터를 지정하여 다양한 종류의 join을 할 수 있음
      - inner join: default, 일치하는 값이 있는 경우 join
      - left: left outer join
      - right: right outer join
      - outer: full outer join
  - `pandas.merge` 함수 이용

```python
customer = pd.DataFrame({'customer_id': np.arange(6), 
                         'name': ['철수', '영희', '길동', '영수', '수민', '동건'], 
                         '나이': [40, 20, 21, 30, 33, 18]})
customer
```

```python
orders = pd.DataFrame({'customer_id': [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
                      'item': ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'],
                       'quantity': [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]
                      })
orders
```

```python
# on -> join 기준이 되는 컬럼 지정, how 파라미터 지정이 없으면 inner join
pd.merge(customer, orders, on='customer_id') 
```

```python
# left outer join: 왼쪽 DataFrame(SQL 테이블)은 그대로 남겨둠(일치하는 값이 없는 경우 NaN)
pd.merge(customer, orders, on='customer_id', how='left') 
```

```python
# right outer join: 오른쪽 DataFrame을 남겨둠
pd.merge(customer, orders, on='customer_id', how='right') 
```

```python
# outer join: DataFrame을 그대로 병합
pd.merge(customer, orders, on='customer_id', how='outer') 
```

```python
merged_data = pd.merge(customer, orders, on='customer_id')
merged_data
```

```python
merged_data.groupby('item').sum().sort_values(by='quantity', ascending=False)
```

```python
merged_data.groupby(['name', 'item']).sum()
```

```python
merged_data.groupby(['name', 'item']).sum().loc['영희', 'quantity']
```
